### Çizgi Grafiği

```python
x = np.arange(-10, 11)
y = x ** 2
x, y
```

```python
plt.plot(x, y)  # Sonuna ";" koyarsanız [<matplotlib.lines.Line2D at 0x27df973fb10>] yazısı kaybolur.
```

```python
plt.plot(x, y)
plt.xlabel('x degerleri')
plt.ylabel('y degerleri')
plt.title('y = $x^{2}$ Eğrisinin Grafiği');
```

*** sort_index(ascending=False) : Tersine sıralama yapmaya yarar.

```python
tur = pd.DataFrame ({'Year':[2023, 2022, 2021, 2020, 2019, 2018, 2017, 2016, 2015],
                    'Rate':[64.27, 97.72, 79.89, 25.15, 7.36, 33.64, 15.47, 9.94, 5.71]})
alm = pd.DataFrame({'Year':[2023, 2022, 2021, 2020, 2019, 2018, 2017, 2016, 2015],
                    'Rate':[3.2, 7.9, 3.1, 0.5, 1.4, 1.5, 2.3, 1.7, 0.1]})
```

Karşılaştırmalı çizgi grafiği görüntüsü elde etmek için;

```python
plt.plot(tur.Year, tur.Rate)
plt.plot(alm.Year, alm.Rate);
```

** Eksen etiketleri ve başlık ekleme;

```python
plt.plot(tur.Year, tur.Rate)
plt.plot(alm.Year, alm.Rate)
plt.xlabel('Yıllar')
plt.ylabel('Oranlar')
plt.title('Enflasyon Karşılaştırma Grafiği');
```

** Grafik etiketleri ekleme;

```python
plt.plot(tur.Year, tur.Rate)
plt.plot(alm.Year, alm.Rate)
plt.xlabel('Yıllar')
plt.ylabel('Oranlar')
plt.title('Enflasyon Karşılaştırma Grafiği')
plt.legend(['Türkiye', 'Almanya']);
```

Şu şekilde de yazabiliriz;

```python
plt.plot(tur.Year, tur.Rate, label = 'Türkiye')
plt.plot(alm.Year, alm.Rate, label = 'Almanya')
plt.xlabel('Yıllar')
plt.ylabel('Oranlar')
plt.title('Enflasyon Karşılaştırma Grafiği')
plt.legend();
```

### Çoklu Çubuk Grafikleri

```python
gunler = ['Pazartesi','Sali','Carsamba','Persembe','Cuma']
A_sirketi = [30, 40, 50, 60, 45]
B_sirketi = [10, 35, 65, 20, 30]
C_sirketi = [20, 15, 30, 40, 65]
```

```python
np.add(A_sirketi, B_sirketi)
```

A_sirketi = np.array([30, 40, 50, 60, 45])
B_sirketi = np.array([10, 35, 65, 20, 30])
C_sirketi = np.array([20, 15, 30, 40, 65]) 

Şeklinde yazmış olsaydık yukarıdaki np.add() işlemini yapmamıza gerek kalmazdı. 

```python
plt.bar(gunler, A_sirketi)
```

```python
plt.bar(gunler, A_sirketi, label= 'A')
plt.bar(gunler, B_sirketi, bottom= A_sirketi, label= 'B')
plt.bar(gunler, C_sirketi, bottom= np.add(A_sirketi, B_sirketi), label= 'C')
plt.legend();
```

```python
genislik = 0.3
xpos = np.arange(5)
```

```python
plt.bar(xpos,A_sirketi,width=genislik, label = 'A Sirketi')
plt.bar(xpos+genislik,B_sirketi,width=genislik, label = 'B Sirketi')
plt.bar(xpos+2*genislik,C_sirketi,width=genislik, label = 'C Sirketi')
plt.xticks(xpos, gunler)
plt.legend();
```

```python
# Yazıları da ortalamak istersek;

plt.bar(xpos,A_sirketi,width=genislik, label = 'A Sirketi')
plt.bar(xpos+genislik,B_sirketi,width=genislik, label = 'B Sirketi')
plt.bar(xpos+2*genislik,C_sirketi,width=genislik, label = 'C Sirketi')
plt.xticks(xpos+genislik, gunler)
plt.legend();
```
