## Koşullu Seçimler

```python
df = pd.DataFrame({'id':[1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008, 1009],
                   'name':['ali', 'ayse', 'fatma', 'murat', 'hasan', 'betul', 'berke', 'nuran', 'nese'],
                   'surname':['guzel', 'can', 'yilmaz', 'kaya', 'demir', 'sahin', 'celik', 'aydin', 'aslan'],
                   'gender':['E', 'K', 'K', 'E', 'E', 'K', 'E', 'K', 'K'],
                   'city':['istanbul', 'izmir', 'ankara', 'istanbul', 'izmir', 'izmir', 'antalya', 'edirne', 'ankara'],
                   'year':[1990, 1995, 1992, 1980, 2000, 2005, 2010, 1999, 1985],
                   'price':[5000, 3000, 2000, 7000, 14000, 6000, 2600, 7500, 8300]})
df
```

```python
kosul = df['city'] == 'izmir'
df[kosul]
```

## Dropna - Eksik Verileri Kaldırmak İçin Kullanılır

```python
np.random.seed(35)
df = pd.DataFrame({
    'Date' : pd.date_range(start='2023-01-13',periods=10,freq='D'),
    'Urun' : 222,
    'Deger_1': np.random.randint(1,10,10),
    'Deger_2': np.random.randint(1,10,10),
    'Deger_3': np.random.randint(1,10,10),
    'Deger_4': np.random.randint(1,10,10)
    
})
df
```

```python
df.loc[[2,6,9],"Urun"] = np.nan
df.loc[[4,6],"Deger_1"] = np.nan
df.loc[[4,6],"Deger_2"] = np.nan
df.loc[6:,"Deger_3"] = np.nan
df.loc[3:6,"Deger_4"] = np.nan
df
```

```python
df.isnull() # Null değerleri gösterir. Ya da df.isna()
```

```python
df.isnull().sum() 
```

** Satırlardaki toplam null değer sayısını döndürür. Toplam null değer sayısını ise kod sonuna .sum() ekleyerek bulabiliriz.

```python
df.count() # Kaç tane null olmayan değer olduğunu döndürür.
```

```python
df.dropna()
```

** İçerisinde en az 1 tane null olan değer olan satırları kaldırır.

df.dropna()
axis=0 : Satırları 
axis=1 : Sütunları kapsar.

```python
df.dropna(axis=1)
```

### Bütün değerleri null olan satırı kaldırmak için;

```python
df.dropna(how= 'all')
```

## Fillna - Eksik Değerleri Doldurma

```python
data = {'isim':['Zeynep','Yusuf','Elif','Ömer','Hamza','Asya','Fatma','Murat','Ahmet','Defne','Can','Berk',np.nan],
       'sirket':['Tesla','Apple','Amazon','Amazon','Tesla','Tesla','Apple','Apple',np.nan,'Amazon','Tesla','Amazon','Apple'],
       'departman':['IT','Sales',np.nan,'Marketing','IT','IT','IT',np.nan,'Marketing','Marketing',np.nan,'Sales','Sales'],
       'maas':[100,150,200,np.nan,np.nan,np.nan,230,np.nan,225,115,160,250,220],
       'prim':[np.nan,np.nan, 10,20,35,34,22,25,40,10,np.nan,35,20]}
df = pd.DataFrame(data)
df
```

#### Aynı değerle doldurma;

```python
df.fillna(0)
```

#### Her sütundaki eksik değeri farklı değerle doldurma;

```python
df.fillna({'isim': 'İsim Belirtilmemiş', 'maas':0, 'departman':'Bilinmiyor', 'prim':0, 'sirket':'Bilinmiyor'})
```

```python
new_df = df.copy()
new_df
```

```python
new_df.ffill() # Kendinden önceki değerle doldurur eksik veriyi.
```

```python
new_df.bfill() # Kendinden sonraki değerle doldurur eksik veriyi.
```
