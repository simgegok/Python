## Aggregate Functions

Aggeration metodları
- count → Tüm elemanların sayısı
- mean → Tüm elemanların ortalamalarını hesaplar
- median → Tüm elemanların median’ı alır
- first, last → İlk ve Son elemanı getirir
- sum → Tüm elemanların toplamlarını alır
- min, max → Tüm elemanların minimum ve maximum değerlerini getirir
- std, var → Tüm elemanların standart sapmasını ve varyansını hesapları
- prod → Tüm elamanların çarpımını hesaplar

```python
data = {'isim':['Zeynep','Yusuf','Elif','Ömer','Hamza','Asya','Fatma','Murat','Ahmet','Defne'],
       'sirket':['Tesla','Apple','Amazon','Amazon','Tesla','Tesla','Apple','Apple','Amazon','Tesla'],
       'departman':['IT','Sales','Marketing','IT','IT','IT','Marketing','Marketing','Sales','Sales'],
       'maas':[100,150,200,125,200,225,100,160,250,220],
       'prim':[10,20,35,34,22,25,40,10,35,20]}

df = pd.DataFrame(data)
df
```

```python
df['maas'].sum()
```

```python
df['maas'].mean()
```

```python
df['maas'].median()
```

```python
df.groupby('sirket').agg({'maas': 'sum'})
```

```python
df.groupby('sirket').agg({'maas': ['sum', 'mean']})
```

```python
df.groupby('sirket').agg({'maas': ['sum', 'mean'], 'prim': ['sum', 'mean']})
```

```python
df.groupby(['sirket', 'departman'])[['maas', 'prim']].sum()
```

```python
df.groupby(['departman', 'sirket'])[['maas', 'prim']].sum()
```
