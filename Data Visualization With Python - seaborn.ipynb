# SEABORN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()  # Seaborn içinde yüklü olan verisetleri 
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Saçılım Grafiği (Scatter Plot)

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')

# Cinsiyete göre etiket rengi verildi.
```

```python
plt.rcParams['figure.figsize'] = (10,4)
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', size='day')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', alpha = .5)
```

### Time Series

```python
df = pd.read_excel('https://github.com/Mylinear/arac_uretim_verisi/raw/main/arac.xlsx')
df
```

```python
df.Tarih = pd.to_datetime(df.Tarih)
```

```python
df
```

```python
df.info()
```

```python
df.columns
```

```python
# İsimlerdeki boşlukları silmek için;

df.columns.str.strip()
```

```python
df.columns = df.columns.str.strip()
```

```python
sns.lineplot(data=df, x='Tarih', y='Binek')
```

```python
df['Ay'] = df.Tarih.dt.month
```

```python
df
```

```python
# Aylara göre üretim değerlerinin ortalamasını alma;

df.groupby('Ay')['Binek'].mean()
```

```python
sns.lineplot(df.groupby('Ay')['Binek'].mean())
```
